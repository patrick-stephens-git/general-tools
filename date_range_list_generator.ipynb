```python
import pandas as pd
from datetime import datetime, date, timedelta
```

```python
length_in_days = 30
```

```python
def date_range_list_generator(length_in_days):
    length_in_days_a = length_in_days
    # print(length_in_days_a)
    length_in_days_b = length_in_days_a + 1
    # print(length_in_days_b)
    end_date_a = date.today()
    # print('         Now: {0}'.format(datetime.now()))
    start_date_a = end_date_a - timedelta(days = length_in_days_a)
    end_date_b = date.today() - timedelta(days = 1)
    start_date_b = end_date_b - timedelta(days = length_in_days_b)
    # print('Start Date A: {0}'.format(start_date_a))
    # print('Start Date B: {0}'.format(start_date_b))
    # print('  End Date A: {0}'.format(end_date_a))
    # print('  End Date B: {0}'.format(end_date_b))
    date_list_a = pd.date_range(start_date_a, end_date_a).tolist()
    # print(date_list_a)
    date_list_b = pd.date_range(start_date_b, end_date_b).tolist()
    # print(date_list_b)
    date_list_b = date_list_b[1:]
    # print(date_list_b)
    date_list_a_clean = []
    for date_a in date_list_a:
        # print(date_a.strftime('%Y-%m-%d'))
        date_list_a_clean.append(date_a.strftime('%Y-%m-%d'))
    # print(date_list_a_clean)
    date_list_b_clean = []
    for date_b in date_list_b:
        # print(date_a.strftime('%Y-%m-%d'))
        date_list_b_clean.append(date_b.strftime('%Y-%m-%d'))
    # print(date_list_b_clean)
    date_range_df = pd.DataFrame({'start_date':date_list_b_clean, 'end_date':date_list_a_clean})
    # date_range_df = date_range_df.sort_values(by='start_date', ascending=False)
    date_range_df = date_range_df.sort_values(by='start_date', ascending=True)
    date_range_df.drop(date_range_df.tail(1).index,inplace=True) # Drop row for "today"
    return date_range_df
```

```python
date_df = date_range_list_generator(length_in_days)
date_df
```
